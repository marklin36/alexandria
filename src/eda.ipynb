| Field name | Format or example  |  Description  |
|---|---|---|
|StoryID |13 unit alpha-numeric code|Unique ID assigned to every news item that ACTA processes in case the underlying text needs to be referred to at a later date.|
|Timestamp|yyyy-mm-dd hh:mm:ss.ZZZ (Z=millisecond)|The publication time stamp of the news item in UTC (Coordinated Universal Time) by default.|
|Ticker|“IBM”|The trading symbol of the company identified in the news item for each listed exchange, point-in-time sensitive|
|Country|“USA”|3-letter country code based on ISO-3166 designation, assigned by the location of the exchange on which the stock is listed, point-in-time sensitive|
|ISIN|“US4592001014”|The International Securities Id Number of the company identified in the news item, point-in-time sensitive|
|Sentiment (trinary score)|1, 0, or -1|Reveals in a trinary score the highest probable state of sentiment found in the news item, 1 = positive, 0 = neutral, and -1 = negative.|
|Confidence|0.000 to 1.000|The probability of the trinary sentiment score (above) matching a known sentiment state alone: 0.000 = zero probability, 1.000 is absolute probability|
|Novelty|(whole number, 1 or >1)|Identifies new stories: 1 = new story, 2 or greater means the item is linked to a previous story that was published within the last 24 hours|
|Event|Prefix AA@ followed by tag, e.g. “AA@ERN”|Identifies the news event / topic of the news item. The tag @ALEX is assigned to each line of data by default, then additional tags will be applied for the most relevant topics found. See News Event list below|
|Relevance|0.1 to 1.0|Weighting of the sentiment to the company in that line of data. Relevance of “1.00”=primary relevance, any less indicates the appropriate degree of relevance. |
|HeadlineOnly|Binary: “T” or “F”|Identifies if the item is headline only or if there is a body as well. “T” = true, for headline only, “F” = false, indicating there is a body to the story.|
|AutoMessage|Binary: “T” or “F”|Identifies if the story is generated as part of a schedule delivery. “T” = true, for the item being recurring/scheduled, “F” = false, indicating that is was unscheduled.|
|Source|3-letter acronym, e.g. “DJN”|A tag that identifies the publication name or source type. See Source ID tab for full list|
|MarketImpactScore|A single factor from -5 to +5|The estimated risk-adjusted 1 minute forward return for a given article as measured by stock return minus the stock beta multiplied market return, -5 being most negative impact and +5 most positive|
|Prob_POS|0.0000 to 1.0000|Sentiment probability: The raw probability that the article sentiment is positive expressed as a decimal with all three probabilities summing to 1|
|Prob_NTR|0.0000 to 1.0000|Sentiment probability: The raw probability that the article sentiment is neutral expressed as a decimal with all three probabilities summing to 1|
|Prob_NEG|0.0000 to 1.0000|Sentiment probability: The raw probability that the article sentiment is negative expressed as a decimal with all three probabilities summing to 1|\n
